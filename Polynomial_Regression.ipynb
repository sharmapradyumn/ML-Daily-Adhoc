```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# data loading
data = pd.read_csv('salary.csv')
```

```python
# exp
exp = data.iloc[:,0:1].values
```

```python
# sal
sal = data.iloc[:,1:].values
```

```python
# call linear regression just for compare
from sklearn.linear_model import LinearRegression
regrl = LinearRegression()
```

```python
# to transfer data we call polynomial transform method
from sklearn.preprocessing import PolynomialFeatures
```

```python
# experince transform
poly=PolynomialFeatures(degree=12)
poly
```

```python
# exp fit
newexp = poly.fit_transform(exp)
```

```python
# apply linear regression
trained1 = regrl.fit(exp,sal)
```

```python
# now creating linear model
plt.xlabel('exp')
plt.ylabel('sal')
plt.scatter(exp,sal,label='actual sal')
plt.plot(exp,trained1.predict(exp),label='linear model predict sal')
plt.legend()
plt.show()
```

```python
# applying new exp data
regr2 = LinearRegression()
```

```python
# new exp data apply
trained2 = regr2.fit(newexp,sal)
```

```python
plt.xlabel('newexp')
plt.ylabel('sal')
plt.scatter(exp,sal,label='actual sal')
plt.plot(exp,trained2.predict(newexp),label='polynomial model predict newsal')
plt.legend()
plt.show()
```
