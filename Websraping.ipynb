```python
from bs4 import BeautifulSoup as bs
```

```python
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Machine_learning'
```

```python
response = requests.get(url)   #sending request and getting response
response
```

```python
# its print all html tags
#response.text
```

```python
#parsing library - html.parser and second is lxml
# its fetch content and html tag both
soup = bs(response.text,"lxml")
```

```python
#soup
```

```python
 for link in soup.find_all('a'):
        #print(link)   # its print link with anker <a> tag
        print(link.get('href')) # only link
        
```

```python
for ptag in soup.find_all('p'):
    #print(ptag)
    #print(ptag.get('b'))
    for btag in ptag.find_all('b'):
        print(btag.text)
    
```
