```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
```

```python
 while  cap.isOpened() :
    status,frame=cap.read()
    # converting frame into gray image 
    grayimg = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    print(frame.shape)
    
    
    cv2.line(frame,(0,0),(100,100),(255,0,0),4)
    cv2.rectangle(frame,(120,130),(300,300),(255,0,0),4)
    cv2.circle(frame,(200,300),100,(13,44,123),2)
    
    
    #cv2.circle(frame,(100,120),50(255,0,0),4)
    #cv2.line(frame,(0,0),(100,100),(255,0,0),4)
    #cap.imshow('live',frame)
    
    
    font = cv2.FONT_HERSHEY_SIMPLEX   #   this  font  type  
    cv2.putText(frame,'PRADYUMN',(10,50), font, 2,(0,2,55),2,cv2.LINE_AA)
    
    
    
    cv2.imshow('live',frame)
    cv2.imshow('livegray',grayimg)   # show gray image
    cv2.imshow('live',frame)
    if  cv2.waitKey(10)   &   0xff  ==  ord('q')  :
        break
cv2.destroyALLWindows()
cap.release()
```
